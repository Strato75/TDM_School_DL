```python
data = [('2011\n(XRCE)', 28.2, 'shallow'), ('2012\n(AlexNet)', 16.4, '8 layers'), 
        ('2013\n(ZF)', 11.7, '8 layers'), ('2014\n(VGG)', 7.3, '19 layers'),
        ('2014\n(GoogLeNet-V1)', 6.7, '22 layers'),
        ('Human${^{(1)}}$', 5.1, ''),
        ('2015\n(ResNet)', 3.6, '152 layers'),
        ('2016\n(GoogLeNet-V4)', 3.1, '22 layers'),
       ]
```

```python
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
figure(figsize(8,5))
bars = bar(x,y)
xticks(fontsize=10, rotation=45)
bars[5].set_color('r')
bars[0].set_color('g')
# Add the data value on head of the bar
for value in bars:
    height = value.get_height()
    plt.text(value.get_x() + value.get_width()/2.,
             1.002*height,'%.1f' % float(height), ha='center', va='bottom')
xlabel("Network name", fontsize=12)
ylabel("Classification error (Top-5 %)", fontsize=12)
title('ImageNet Large Scale Visual Recognition Challenge (ILSVRC)', fontsize=12)
ylim(0,30)
tight_layout()

savefig('figures/ILSVRC.png', dpi=300)
```
