```python
un_papers = {'Berkeley': 33, 
'Cmu': 32,
'Stanford': 29,
'Mit': 28,
'Oxford': 20}

inst_papers = {'Google': 43,
'Deepmind': 33,
'Microsoft': 28,
'Google AI': 24,
'Facebook': 21}
```

```python
# universities
data = un_papers

x = [i for i in data]
y = [data[i] for i in data]

figure(figsize=(8,5))
bar(x,y)
xticks(fontsize=12)
yticks(fontsize=12)
xlabel('Univeristy')
ylabel('Accepted papers')
title('International Conference on Machine Learning (ICML) 2018')

savefig('figures/ICML2018_un.png', dpi=300)
```

```python
# universities
data = inst_papers

x = [i for i in data]
y = [data[i] for i in data]

figure(figsize=(8,5))
bar(x,y)
xticks(fontsize=12)
yticks(fontsize=12)
xlabel('Commercial insitution')
ylabel('Accepted papers')
title('International Conference on Machine Learning (ICML) 2018')
savefig('figures/ICML2018_inst.png', dpi=300)
```

```python
# universities
data0 = un_papers
data1 = inst_papers

x0 = [i for i in data0] + [i for i in data1]
y0 = [data0[i] for i in data0] + [data1[i] for i in data1]


figure(figsize=(8,5))
bars = bar(x0,y0)
for i in range(5):
    bars[i].set_color('r')
    
for i in range(5, 10):
    bars[i].set_color('b')

xticks(fontsize=12, rotation=45)
yticks(fontsize=12)
xlabel('Univerist e aziende', fontsize=12)
ylabel('Articoli accettati', fontsize=12)
title('International Conference on Machine Learning (ICML) 2018', fontsize=14)
tight_layout()

savefig('figures/ICML2018.png', dpi=300)
```
